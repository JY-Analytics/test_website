# Spotify Playlist Comparison


### An Analysis of My Favorite Songs in 2020 vs 2021

# Project Overview

For this project, I chose to analyze my how my own music taste has evolved in the past year.

Spotify is a music streaming application, which allows the user to stream or download many popular artists. It has a wide array of functionality, including allowing the user to create playlists.

Spotify has one special feature known as "Your Top Songs". Each year, Spotify will automatically create a new playlist of the user's top 100 songs, which consists of the songs the user listened to the most during the calendar year.

In this notebook, I will compare my Top Songs for 2020 and 2021 to see how my music taste has changed.

# Loading the Data

The following website allows Spotify users to downloand .csv files of their Spotify playlists:
https://exportify.net/

After downloading all playlists, I saved the two .csv files for 2020 and 2021 in this workbook environment.

```python
import pandas as pd
from io import StringIO
from IPython.display import display
```

```python
# Import data: One .csv file for each playlist
data_2020 = pd.read_csv('your_top_songs_2020.csv')
data_2021 = pd.read_csv('your_top_songs_2021.csv')

print("=== 2020 data set: {} rows x {} columns. ===".format(data_2020.shape[0], data_2020.shape[1]))
display (data_2020.head())

print("=== 2021 data set: {} rows x {} columns. ===".format(data_2021.shape[0], data_2021.shape[1]))
display (data_2021.head())
```

# Cleaning the Data

To work with the data, the 2 data sets should be combined. We will want to preserve the playlist year, as well as the index of each song in its own playlist. The index in this case has special meaning: The top tracks had more plays.

I will also create a SQL database table of the data to make querying easier.

```python
# Create a Year field before combining the data
data_2020["Year"] = 2020
data_2021["Year"] = 2021

# Preserve the Song's rank before combining the data
data_2020['Rank'] = data_2020.index + 1 #add one for conventional ranking: top song == #1
data_2021['Rank'] = data_2021.index + 1

# Combine the datasets
df = pd.concat([data_2020, data_2021])

# Delete irrelevant fields
df = df.drop(df.columns[[0, 2, 4, 6, 9, 10, 13, 16, 17]], axis=1)

# Fix column names to simplify them
df = df.rename(columns={df.columns[0]: "Track", 
                        df.columns[1]: "Artist", 
                        df.columns[2]: "Album", 
                        df.columns[3]: "Album Artist",
                        df.columns[4]: "Release Date", 
                        df.columns[5]: "Track Number", 
                        df.columns[6]: "Duration"})

# Convert Release Date to datetime
df['Release Date'] = pd.to_datetime(df['Release Date'])

# Convert miliseconds into seconds
df["Duration"] = df["Duration"].apply(lambda x: int(x/1000))


display(df)
```

```python
# Setup a SQL data table to run queries on

import pandas as pd
import sqlite3

conn = sqlite3.connect('music_db')
c = conn.cursor()

c.execute('CREATE TABLE IF NOT EXISTS music (track text, artist text, album text, album_artist text, album_release date, track_number number, track_time number, explicit number, popularity number, year date, rank number)')
conn.commit()

df.to_sql('music', conn, if_exists='replace', index = False)

```

# Analysis

Did any songs make the top 100 in for both years?

```python
query = '''
  SELECT [Track], [Artist], COUNT(*), AVG([Rank])
  FROM music
  GROUP BY [Track]
  HAVING COUNT(*) > 1
  ORDER BY [Artist]
'''

df2 = pd.read_sql_query (query, conn)

display(df2)
```

Which artists appeared more than once?

```python
query = '''
  SELECT [Artist], AVG([Rank]), COUNT(*)
  FROM music
  GROUP BY [Artist]
  HAVING COUNT(*) > 1
  ORDER BY COUNT(*) DESC
'''

df2 = pd.read_sql_query (query, conn)

display(df2)
```

How did the average duration, average popularity, and portion of explicit songs change between 2020 and 2021?

```python
query = '''
  SELECT [Year], AVG([Duration]), AVG([Popularity]), SUM([Explicit]), AVG([Release Date])
  FROM music
  GROUP BY [Year]
  ORDER BY COUNT(*) DESC
'''

df2 = pd.read_sql_query (query, conn)

display(df2)
```

To sumarize the above:

I listend to more popular music and more music with explicit lyrics.

My songs were slightly shorter on average. However the average song length only decreased by 1 second. A typical song in my top 100 is 3 minutes and 41 seconds.

As one would expect, the average album release date was slightly newer.

## Comparing Song Popularity to My Rank

One potential hypothesis is that I am more likely to rank more popular artists higher than less popular ones. We could investigate this visually.

```python
query = '''
  SELECT [Artist], AVG([Rank]) As Rank, AVG([Popularity]) As Popularity
  FROM music
  GROUP BY [Artist]
  ORDER BY AVG([Popularity]) DESC
'''

df_pop = pd.read_sql_query (query, conn)

display(df_pop)
```

```python
df_pop.plot.scatter(x = 'Popularity', y = 'Rank')
```

The hypothesis appears to be incorrect. Popularity does not predict how much I will enjoy an artist.

As a bonus, the table above shows the most popular and least popular artists on my playlists.
